```python
import pygrt
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
mod = "halfspace2"
depsrc = "0.10000"
deprcv = "0.00000"
strdists = [p.split("_")[-1] for p in glob.glob(f"stgrtstats/{mod}_stats_{depsrc}_{deprcv}/K_*")]
strdists.sort()
sdist = strdists[22]
print(sdist)
data = pygrt.utils.read_statsfile(f"stgrtstats/{mod}_stats_{depsrc}_{deprcv}/K_{sdist}")

fig, ax = pygrt.utils.plot_statsdata(data, "DC", "2", "w", "2")
```

```python
ptamdata = pygrt.utils.read_statsfile_ptam(f"stgrtstats/{mod}_stats_{depsrc}_{deprcv}/PTAM_{sdist}")
fig, ax = pygrt.utils.plot_statsdata_ptam(data, ptamdata, "DC", "2", "3")
```
