```python
import numpy as np
import matplotlib.pyplot as plt
```

compare displacements

```python
dipslip1 = np.loadtxt("syn_1_dipslip")
dipslip2 = np.loadtxt("syn_2_dipslip")
stkslip1 = np.loadtxt("syn_1_stkslip")
stkslip2 = np.loadtxt("syn_2_stkslip")

fig, axs = plt.subplots(4, 3, figsize=(10,12))
for i, slip in enumerate([stkslip1, dipslip1, stkslip2, dipslip2]):
    ax3 = axs[i]
    chs = ['Z', 'N', 'E']
    for k in range(3):
        ax = ax3[k]
        y = slip[:, k+2].copy()
        if chs[k] == 'E':
            y *= -1
        ax.plot(slip[:, 1], y, label=chs[k])
        ax.grid()
        ax.xaxis.set_inverted(True)
        ax.legend()
```

compare stress

```python
dipslip1 = np.loadtxt("stress_1_dipslip")
dipslip2 = np.loadtxt("stress_2_dipslip")
stkslip1 = np.loadtxt("stress_1_stkslip")
stkslip2 = np.loadtxt("stress_2_stkslip")

fig, axs = plt.subplots(4, 2, figsize=(8,12))
for i, data in enumerate([stkslip1, dipslip1, stkslip2, dipslip2]):
    ax3 = axs[i]

    ax = ax3[0]
    ax.plot(data[:,1], data[:, 2], label='ZZ', c='b')
    ax.plot(data[:,1], data[:, 5], label='NN', c='k')
    ax.plot(data[:,1], data[:, 7], label='EE', c='r')
    ax.grid()
    ax.xaxis.set_inverted(True)
    ax.legend()

    ax = ax3[1]
    ax.plot(data[:,1], data[:, 3], label='ZN', c='r')
    ax.plot(data[:,1], -data[:, 4], label='ZE', c='b')
    ax.plot(data[:,1], -data[:, 6], label='NE', c='k')
    ax.grid()
    ax.legend()
    ax.xaxis.set_inverted(True)
```
