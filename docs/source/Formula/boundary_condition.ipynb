```python
import sympy as sp
```

## 动态解

```python
# 定义基本变量
k, b, d, a, mu, Omg, kb, ka, w, Delta = sp.symbols(r'k b d, a mu \Omega k_{b} k_{a} \omega \Delta')
lam, rho = sp.symbols(r'\lambda \rho')
```

```python
D = sp.Matrix([
    [k,            b,             k,            -b],
    [a,            k,             -a,           k],
    [2*mu*Omg,     2*k*mu*b,      2*mu*Omg,     -2*k*mu*b],
    [2*k*mu*a,     2*mu*Omg,      -2*k*mu*a,    2*mu*Omg]
])
```

```python
def get_R(D, at_bottom:bool):
    if not at_bottom:
        return - (D[:, 2:])**(-1) * (D[:, :2])
    else:
        return - (D[:, :2])**(-1) * (D[:, 2:])
```

### 自由界面（z平面牵引力为0）

```python
# z+ 侧， 例如顶层顶界面
R = get_R(D[2:, :], False)
R.simplify()
R
```

```python
# z- 侧， 例如底层底界面
R = get_R(D[2:, :], True)
R.simplify()
R
```

### 刚性界面（位移为0）

```python
# z+ 侧
R = get_R(D[:2, :], False)
R.simplify()
R
```

```python
# z- 侧
R = get_R(D[:2, :], True)
R.simplify()
R
```

## 静态解

```python
# 在静态解中，矩阵 D 和相对深度 z-zj 有关, 因此 z+ 侧和 z- 侧需使用两种 D 矩阵
D1 = sp.Matrix([
    [1,         -(1 + 2*k*Delta*d),            1,            -(1 - 2*k*Delta*d)],
    [1,          (1 - 2*k*Delta*d),           -1,            -(1 + 2*k*Delta*d)],
    [2*mu*k,     2*mu*Delta*k*(1 - 2*k*d),   2*mu*k,      2*mu*Delta*k*(1 + 2*k*d)],
    [2*mu*k,    -2*mu*Delta*k*(1 + 2*k*d),  -2*mu*k,      2*mu*Delta*k*(1 - 2*k*d)],
])

z = 0
D2 = sp.Matrix([
    [1,         -(1 + 2*k*Delta*z),            1,            -(1 - 2*k*Delta*z)],
    [1,          (1 - 2*k*Delta*z),           -1,            -(1 + 2*k*Delta*z)],
    [2*mu*k,     2*mu*Delta*k*(1 - 2*k*z),   2*mu*k,      2*mu*Delta*k*(1 + 2*k*z)],
    [2*mu*k,    -2*mu*Delta*k*(1 + 2*k*z),  -2*mu*k,      2*mu*Delta*k*(1 - 2*k*z)],
])
```

### 自由界面（z平面牵引力为0）

```python
# z+ 侧
R = get_R(D2[2:, :], False)
R.simplify()
R
```

```python
# z- 侧
R = get_R(D1[2:, :], True)
R.simplify()
R
```

### 刚性界面（位移为0）

```python
# z+ 侧
R = get_R(D2[:2, :], False)
R.simplify()
R
```

```python
# z- 侧
R = get_R(D1[:2, :], True)
R.simplify()
R
```
